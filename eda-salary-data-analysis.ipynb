## Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Load the Dataset

```python
df=pd.read_csv('/kaggle/input/salary-data/Salary_Dataset_DataScienceLovers.csv')
df.head()
```

## Basic Exploration

```python

df.info()
```

```python
df.describe()
```

## Data Cleaning

### 1) Check Missing Values 

```python
df.isnull().sum()
```

```python
#Remove the single row of null
df=df.dropna(subset=['Company Name'])
```

### 2) Removing Duplicate Rows

```python
df.duplicated().sum()
#np.int64(0) means there are 0 duplicate rows in dataset,hence no need for drop
```

### 3) Outlier detection and Removal

#### Visualize Salary Outliers

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Salary'])
plt.title("Salary Outliers")
plt.show()
```

```python
# IQR(Interquartile Range) Method
Q1=df['Salary'].quantile(0.25)
Q3=df['Salary'].quantile(0.75)
IQR=Q3-Q1

lower_limit=Q1 - 1.5  * IQR
upper_limit=Q3 + 1.5 * IQR

df=df[(df['Salary']>=lower_limit) & (df['Salary']<=upper_limit)]
df
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Rating'])
plt.title("Rating Outliers")
plt.show()
```

### Box plot analysis shows that most company rating lie between 3.5-4.5. Lower rating appear as statistical outliers but are valid observations and were retained for analysis

## Employment Status Offers the Highest Salaries 
Full‑time employees earn the most, while interns and trainees earn significantly less.

```python
# Salary share by Employment Status
salary_by_status = df.groupby('Employment Status')['Salary'].mean()

plt.figure(figsize=(6,6))
plt.pie(salary_by_status, labels=salary_by_status.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Average Salary Share by Employment Status")
plt.show()

```

```python
## Salary Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Salary'],bins=30,kde=True)
plt.ticklabel_format(style='plain',axis='x')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```

```python
# Rating share by Job Roles
rating_by_role = df.groupby('Job Roles')['Rating'].mean()

plt.figure(figsize=(8,8))
plt.pie(rating_by_role, labels=rating_by_role.index, autopct='%1.1f%%', colors=sns.color_palette("Set3"))
plt.title("Average Rating Share by Job Role")
plt.show()
```

## Top 10 Job Roles by Average Salary



```python
#Compute average salary by job role
avg_salary_by_role = df.groupby('Job Roles')['Salary'].mean().sort_values(ascending=False).head(10)

# Step 4: Plot sorted bar chart
plt.figure(figsize=(12,6))
avg_salary_by_role.plot(kind='bar')
plt.title("Top 10 Job Roles by Average Salary")
plt.show()
```

## Average Salary by Location

```python
# Compute average salary by Location (Top 10)
avg_salary_by_location = df.groupby('Location')['Salary'].mean().sort_values(ascending=False).head(10).reset_index()

# Bubble chart
fig = px.scatter(avg_salary_by_location, 
                 x='Location', 
                 y='Salary', 
                 size='Salary', 
                 color='Salary',
                 hover_name='Location',
                 title="Top 10 Locations by Average Salary (Bubble Chart)")

fig.update_xaxes(type='category')
fig.show()
```

## Job title having highest number of salaries reported

```python
df['Job Title'].value_counts()
```

```python
sal_reported=df.groupby('Job Title')['Salaries Reported'].count().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6))
sal_reported.head(5).plot(kind='barh', color='purple', edgecolor='black')
plt.title("Top 10 Job Titles by Number of Salaries Reported")
plt.xlabel("Count of Salaries Reported")
plt.ylabel("Job Title")
plt.gca().invert_yaxis()  # highest at top
plt.show()
```

## Average Salary Trend Across Company Ratings

```python
rating_salary=df.groupby('Rating')['Salary'].mean().reset_index()
#We use reset_index to convert to dataframe
rating_salary.head(10)
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x='Rating',y='Salary',data=rating_salary,marker='o')
plt.xlabel('Company Rating')
plt.ylabel('Average Salary')
plt.title('Average Salary Trend Across Company Ratings')
plt.show()
```

## Mini Dashboard

```python
rating_salary=df.groupby('Rating')['Salary'].mean().reset_index()
top_locations=df.groupby('Location')['Salary'].mean().sort_values(ascending=False)

#Create Dashboard Layout
plt.figure(figsize=(14,10))

#1. Salary Distribution
plt.subplot(2,2,1)
sns.histplot(df['Salary'],bins=30,kde=True)
plt.title("Slary Distribution")

#2. Rating vs Salary
plt.subplot(2,2,2)
sns.scatterplot(x='Rating',y='Salary',data=df,alpha=0.5)
sns.regplot(x='Rating',y='Salary',data=df,scatter=False,color='red')
plt.title('Rating vs Salary')

#Average Salary by Rating
plt.subplot(2,2,3)
sns.lineplot(x='Rating',y='Salary',data=rating_salary,marker='*')
plt.title('Average Salary by Rating')

# Top Locations
plt.subplot(2,2,4)
top_locations.sort_values().plot(kind='barh',color='orange')
plt.title('Top 10 location by Average Salary ')

plt.suptitle("Salary Data Analysis Dashboard",fontsize=15,fontweight="bold")
plt.show()

```
